```python
import numpy as np
import matplotlib.pyplot as plt

#derivadas
def f(x):
  return (x**3)*np.sin(x)

def der_analitica(x):
  return (3*(x**2))*np.sin(x)+(x**3)*np.cos(x)

def der_dir(x,delx):
  return (f(x+delx)-f(x))/delx

def der_ctr(x,delx):
  return (f(x+delx)-f(x-delx))/(2*delx)

#erros
def erro_dir(x,delx):
  return abs(der_analitica(x)-der_dir(x,delx))/abs(der_analitica(x))

def erro_ctr(x,delx):
  return abs(der_analitica(x)-der_ctr(x,delx))/abs(der_analitica(x))

#resultados
result_der_analitica = np.full(shape=50,fill_value=der_analitica(3), dtype=int) #transformar em array se nada der certo
delx = np.logspace(-16, 0, num=50)
array_der_dir = np.array(der_dir(3,delx))
array_der_ctr = np.array(der_ctr(3,delx))
array_err_dir = np.array(erro_dir(3,delx))
array_err_ctr = np.array(erro_ctr(3,delx))

derivadas = plt.figure()

plt.title("Comparação Entre Derivadas")
plt.ylabel("$f'(3)$")
plt.xlabel('$\Delta x$')
plt.xscale("log")
plt.plot(delx,array_der_dir, label='Direita', color='red')
plt.plot(delx,array_der_ctr, label='Central',color='orange')
plt.plot(delx,result_der_analitica,label='Analítica',color='blue')
plt.legend(title = "Derivadas")
plt.savefig('Comparaçao Derivadas.png')
plt.show()

erros = plt.figure()
plt.title("Comparação dos Erros $vs$ Derivada Analítica")
plt.ylabel("Erro para $f'({x})$")
plt.xlabel('$\Delta x$')
plt.plot(delx,array_err_dir,label='Direita',color='red')
plt.plot(delx,array_err_ctr,label='Central', color='orange')
plt.legend(title = "Erros")
plt.xscale("log")
plt.savefig('Comparaçao Erro.png')
plt.show()
```
