#Atividade método do trapézio



```
def int_trapezio_dados(f_lista,dx):
  """
  Função que calcula uma integral de maneira mais simples possível, usando 
  'fatias' trapezoidais. 

  f_lista: lista ou array de valores de uma função avaliada em valores de x 
          separados por intervalo dx. 
  dx: intervalo entre os pontos onde a função foi avaliada

  """
  return dx*(0.5*f_lista[0]+0.5*f_lista[-1]+sum(f_lista[1:-1]))
```

```
import math
import numpy as np

dx=0.0001
x = np.arange(0,math.pi+dx,dx)

f_dados = np.sin(x)

print(int_trapezio_dados(f_dados,dx))
```

#Atividade prática cinemática


```
import numpy as np
import matplotlib.pyplot as plt

t,v=np.loadtxt('velocidades.dat', unpack=True)
```

```
plt.plot(t,v)
plt.xlabel('t(s)')
plt.ylabel('v(m/s)')
plt.show()
```

```
posicao = []

for j in range(1,len(t)):

#a lista posição terá um valor a menos do que a velocidade, pois 
# não podemos integrar de 0 até 0. 

#  a = t[0]
#  b = t[j]  
#  N=j
#  dt = (b-a)/N

  dt = 1.

  f_val=v[0:j]
  posicao.append(int_trapezio_dados(f_val,dt))



figura = plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.plot(t,v)
plt.ylabel('v(m/s)')
plt.xlabel('t(s)')
plt.grid()
plt.title('Velocidade')

plt.subplot(122)
plt.plot(t[1:101],posicao)  #t[1:101] para levar em conta que há um valor a menos de posição do que na lista de tempo
plt.ylabel('x(m)')
plt.xlabel('t(s)')
plt.grid()
plt.title('Posição')

plt.show()
```
